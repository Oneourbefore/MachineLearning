```python
from sklearn.datasets import load_iris 
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

iris.JPG

dataframe.JPG

```python
iris = datasets.load_iris()
```

```python
X_DF = pd.DataFrame(iris.data, columns = iris.feature_names)
X_DF
```

```python
Y_DF = pd.DataFrame(iris.target, columns = ['Target'])
Y_DF
```

```python
X = iris.data[:, [2, 3]] # petal length, petal width 
Y = iris.target # label, class
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
Naive_Bayes = GaussianNB()
```

```python
Y_pred = Naive_Bayes.fit(X_train, Y_train).predict(X_test)
print('Pred : ',Y_pred)
print('True : ',Y_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
cf = confusion_matrix(Y_test, Y_pred)

cf_DF = pd.DataFrame(cf, columns = ['True_0', 'True_1', 'True_2'])
cf_DF.index=['Pred_0', 'Pred_1', 'Pred_2']
cf_DF
```
