```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

X = titanic[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
Y = titanic[['Survived']]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 9)
```

트레인 셋, 테스트셋을 나눴음 변수가 먼저 들어가고 맞춰야할 값을 넣어줌, 테스트사이즈는 전체 데이터에서 30%를 테스트 데이터로 쓰겠다는 얘기임 ( 학습에 전혀 사용하지 않고 테스트에만 사용할 데이터)
random_state를 설정하지 않으면 무작위로 나눔, 이렇게 설정을 하면 이 난수에 해당하는 값에 대해서?

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
prediction = model.predict(X_test) #예측한 값을 출력한 것 
prediction
```

```python
metrics.confusion_matrix(Y_test, prediction)

# True Negative   |  False Positive
# False Negative  |  True Positive
```

```python
print('Accuracy : {:.2f}%'.format(100 * metrics.accuracy_score(Y_test, prediction)))
print('Precision : {:.2f}%'.format(100 * metrics.precision_score(Y_test, prediction)))
print('Recall : {:.2f}%'.format(100 * metrics.recall_score(Y_test, prediction)))
print('F1-score : {:.2f}%'.format(100 * metrics.f1_score(Y_test, prediction)))
```

```python
#cut off 조절에 따른 모델의 성능 평가

metric = []
cut_off_list = [i for i in np.linspace(0,1,100)]

for i in cut_off_list:
    pred = model.predict_proba(X_test)[:,1]>i               # 생존일 확률이 i보다 큰 경우 생존으로 분류
    cf_mtx = metrics.confusion_matrix(Y_test, pred)         # cut off에 따른 confusion matrix
    acc = metrics.accuracy_score(Y_test, pred)              # TP + TN / 전체
    TPR = cf_mtx[1, 1] / (cf_mtx[1,1] + cf_mtx[1,0])        # TP / TP + FN
    FPR = cf_mtx[0, 1] / (cf_mtx[0,1] + cf_mtx[0,0])        # FP / FP + TN
    F1 = metrics.f1_score(Y_test, pred)
    metric.append([i, acc, F1, TPR, FPR])

cut_off = pd.DataFrame(metric)
cut_off.columns = ['Cut off', 'Accuracy', 'F1_socre', 'TPR', 'FPR']
cut_off
```

```python
plt.plot(cut_off["FPR"],cut_off["TPR"], color="darkorange", label="ROC curve")
plt.plot([0,1], [0,1],color='navy', linestyle='--')
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 9)
```
